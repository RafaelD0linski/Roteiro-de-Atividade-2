```python
### Exercício 2 (3,0pts): Use .describe() para examinar as sínteses estatísticas das características de valor da fatura. Reflita sobre o que viu. Faz sentido?
```

```python
import pandas as pd

df = pd.read_excel(r"C:\Users\rafae\Downloads\default_of_credit_card_clients__courseware_version_1_21_19 (1).xls")

bill_amt_features = ["BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]

bill_amt_stats = df[bill_amt_features].describe()
print(bill_amt_stats)
```

```python
### O .describe() fornece uma visão geral das estatísticas descritivas, como média, desvio padrão, mínimo, máximo e quartis. Isso ajuda a entender a distribuição dos valores das faturas. Por exemplo, se a média for muito diferente da mediana, isso pode indicar uma distribuição assimétrica.
```

```python
### Exercício 3 (4,0pts): Visualize as características de valor da fatura usando uma grade 2 por 3 de plotagens de histograma.
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, feature in enumerate(bill_amt_features):
    ax = axes[i//3, i%3]
    df[feature].hist(bins=20, ax=ax)
    ax.set_title(feature)
    ax.set_xlabel('Valor da Fatura')
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
### Exercício 4 (4,0pts): Obtenha o resumo de .describe() para as características de valor do pagamento. Faz sentido?
```

```python
pay_amt_stats = df[pay_amt_features].describe()
print(pay_amt_stats)
```

```python
### Exercício 5 (4,0pts): Plote um histograma das características de pagamento da fatura semelhante ao das características de valor da fatura, mas aplique também alguma rotação aos rótulos do eixo x.
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, feature in enumerate(pay_amt_features):
    ax = axes[i//3, i%3]
    df[feature].hist(bins=20, ax=ax)
    ax.set_title(feature)
    ax.set_xlabel('Valor do Pagamento', rotation=45)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
### Exercício 6 (4,0pts): Use uma máscara booleana para ver quantos dos dados de valor do pagamento são exatamente iguais a 0.
```

```python
zero_payments_mask = df[pay_amt_features] == 0

zero_payments_count = zero_payments_mask.sum()
print(zero_payments_count)
```

```python
### Exercício 7 (4,0pts): Ignorando os pagamentos iguais a 0 usando a máscara que criou na etapa anterior, utilize o método .apply() do pandas e o método np.log10() do NumPy para plotar histogramas de transformações logarítmicas dos pagamentos diferentes de zero.
```

```python
import numpy as np

log_payments = df[pay_amt_features][~zero_payments_mask].apply(np.log10)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for i, feature in enumerate(pay_amt_features):
    ax = axes[i//3, i%3]
    log_payments[feature].hist(bins=20, ax=ax)
    ax.set_title(feature)
    ax.set_xlabel('Log10 do Valor do Pagamento', rotation=45)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```
